Install packages and library

```r
install.packages('tidyverse')
library(tidyverse)
```

Load the CSV into R

```r
daily_activity <- read.csv("dailyActivity_merged.csv")
sleep_day <- read.csv("sleepDay_merged.csv")
```

Exploring the table

```r
head(daily_activity)
colnames(daily_activity)
head(sleep_day)
colnames(sleep_day)
```

Determine how many user in the dataset

```r
n_distinct(daily_activity$Id)
n_distinct(sleep_day$Id)
```

Determine the number of observations in the table

```r
nrow(daily_activity)
nrow(sleep_day)
```

Explore the statistical summary of the data

```r
daily_activity %>%  
  select(TotalSteps,
         TotalDistance,
         SedentaryMinutes) %>%
  summary()
```

```r
sleep_day %>%  
  select(TotalSleepRecords,
  TotalMinutesAsleep,
  TotalTimeInBed) %>%
  summary()
```

Plot the relationship between steps taken in a day and sedentary minutes

```r
ggplot(data=daily_activity, aes(x=TotalSteps, y=SedentaryMinutes)) + geom_point()
```

Plot the relationship between minutes asleep and time in bed

```r
ggplot(data=sleep_day, aes(x=TotalMinutesAsleep, y=TotalTimeInBed)) + geom_point()
```

Merging the dataset

```r
combined_data <- merge(sleep_day, daily_activity, by="Id")
```
